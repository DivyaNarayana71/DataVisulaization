```python
def regionAndMonth(region, month):
    monthName = ''
    if month == 1:
        monthName ='Actual Rainfall: JUN'
    elif month == 2:
        monthName = 'Actual Rainfall: JUL'
    elif month == 3:
        monthName = 'Actual Rainfall: AUG'
    elif month == 3:
        monthName = 'Actual Rainfall: SEPT'
        
    if region == 1:
        print(cIndiaRain[cIndiaRain[monthName]==cIndiaRain[monthName].max()][['YEAR', monthName]])
    elif region == 2:
        print(neIndiaRain[neIndiaRain[monthName]==neIndiaRain[monthName].max()][['YEAR', monthName]])
    elif region == 3:
        print(nwIndiaRain[nwIndiaRain[monthName]==nwIndiaRain[monthName].max()][['YEAR', monthName]])
    elif region == 4:
        print(sIndiaRain[sIndiaRain[monthName]==sIndiaRain[monthName].max()][['YEAR', monthName]])

print("Enter region And month")
print("For Central India Enter 1")
print("For North East India Enter 2")
print("For North West India Enter 3")
print("For South India Enter 4")
region = int(input("Enter the region number:"))

print("\n\nFor june Enter 1")
print("For July Enter 2")
print("For august 3")
print("For september 4")
month = int(input("Enter the month number:\n\n"))
regionAndMonth(region, month)
```
